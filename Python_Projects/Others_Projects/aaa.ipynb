```python
import tkinter as tk
root = tk.Tk()
root.title("ثبت اطلاعات")
root.geometry("600x600")
entry = tk.Entry(root, width=60, font=("B Nazanin", 18))
entry.pack()
def sign():
    content = entry.get().strip()
    with open("data.txt", "w", encoding="utf-8") as file:
        file.write(content)
btn = tk.Button(root, text="ثبت اطلاعات", bg="lightblue", font=("B Titr", 18), command=sign)
btn.pack()
root.mainloop()
```
